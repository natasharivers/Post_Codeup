```python
import numpy as np 
import matplotlib.pyplot as plt 

plt.figure(figsize=(20,8))
X = ['San Antonio Pride Parade','The San Antonio LGBTQ Community Statistics','Podcast Listening: Alok Vaid-Menon','Inclusivity & Diversity Dinner w/ Ben Peavy', 'PRIDE Kick-Off Mixer']
Attendees = [0,17,16,12,40]
RSVPs = [21,31,20,13,26]
  
X_axis = np.arange(len(X))

plt.bar(X_axis - 0.1, Attendees, 0.2, label = '# of Attendees')
plt.bar(X_axis + 0.1, RSVPs, 0.2, label = '# of RSVPs')
  
plt.xticks(X_axis, X)
plt.xlabel("Events")
plt.ylabel("Number of People")
plt.title("Distribution by Event")
plt.legend()
plt.show()
```
