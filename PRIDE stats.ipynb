```python
import numpy as np 
import matplotlib.pyplot as plt 

plt.figure(figsize=(20,8))
X = ['San Antonio Pride Parade','The San Antonio LGBTQ Community Statistics','Podcast Listening: Alok Vaid-Menon','Inclusivity & Diversity Dinner w/ Ben Peavy', 'PRIDE Kick-Off Mixer']
Attendees = [54,18,18,11,33]
RSVPs = [21,31,20,13,26]
  
X_axis = np.arange(len(X))

plt.bar(X_axis - 0.1, Attendees, 0.2, label = '# of Attendees')
plt.bar(X_axis + 0.1, RSVPs, 0.2, label = '# of RSVPs')
  
plt.xticks(X_axis, X)
plt.xlabel("Events")
plt.ylabel("Number of People")
plt.title("Distribution by Event")
plt.legend()
plt.show()
```

```python
# importing library
import matplotlib.pyplot as plt
  
# function to add value labels
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i], y[i], ha = 'center')
  
if __name__ == '__main__':
    
    # creating data on which bar chart will be plot
    x = ['San Antonio Pride Parade','The San Antonio LGBTQ Community Statistics','Podcast Listening: Alok Vaid-Menon','Inclusivity & Diversity Dinner w/ Ben Peavy', 'PRIDE Kick-Off Mixer']
    y = [54,18,18,11,33]
    c = ['red', 'orange', 'yellow', 'green', 'blue']
      
    # setting figure size by using figure() function 
    plt.figure(figsize = (20, 8))
      
    # making the bar chart on the data
    plt.bar(x, y,color = c)
      
    # calling the function to add value labels
    addlabels(x, y)
      
    # giving title to the plot
    plt.title("Pride Event Attendance")
      
    # giving X and Y labels
    plt.xlabel("Event")
    plt.ylabel("Number of Attendees")
       
    # visualizing the plot
    plt.show()
```
